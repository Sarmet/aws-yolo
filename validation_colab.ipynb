```python
# 2. Importăm modulele și încărcăm modelul
import glob
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
%matplotlib inline

model_path = 'best.pt'
if os.path.exists(model_path):
    print(f"Încărcăm modelul din {model_path}...")
    model = YOLO(model_path)
    print("Model încărcat cu succes!")
else:
    print("EROARE: Nu am găsit fișierul best.pt. Verifică dacă clonarea a funcționat.")
```

```python
# 3. Rulăm detecția pe imaginile de test (Validation Samples)
test_images = glob.glob('validation_samples/*.jpg')

if not test_images:
    print("Nu am găsit imagini în folderul validation_samples.")
else:
    print(f"Testăm pe {len(test_images)} imagini...")
    # Rulăm predicția
    results = model(test_images, conf=0.15)

    # Afișăm rezultatele
    for i, r in enumerate(results):
        plt.figure(figsize=(10, 8))
        plt.imshow(cv2.cvtColor(r.plot(), cv2.COLOR_BGR2RGB))
        plt.title(os.path.basename(test_images[i]))
        plt.axis('off')
        plt.show()
```
